# FaceSwap
- OpenCV와 dlib를 사용한 모델입니다.
- 소스 이미지와 타겟 이미지, 총 2장의 이미지가 필요합니다.
- 소스 이미지의 얼굴을 타겟 이미지에 붙여줍니다.
- 출처 : https://github.com/wuhuikai/FaceSwap


```
# 깃클론을 해줍니다.
!git clone https://github.com/wuhuikai/FaceSwap.git
```

```
# 필요한 라이브러리를 가져옵니다.
from PIL import Image
from IPython.display import Image
```

## 1. Choose the source image

```
# 소스 이미지를 업로드 합니다.
%cd /content/FaceSwap/imgs

from google.colab import files
source_img = files.upload()

%cd /content/FaceSwap
```

## 2. Choose the target image

```
# 타겟 이미지를 업로드합니다.
%cd /content/FaceSwap/imgs

from google.colab import files
target_img = files.upload()

%cd /content/FaceSwap
```

## 3. Start translation

```
# 소스 이미지의 얼굴을 타겟 이미지에 붙여줍니다.
import os 

result_name = 'result.jpg'

for source_name in source_img.keys():
    print('source_name : ' + source_name)
    break
for target_name in target_img.keys():
    print('target_name : ' + target_name)
    break

command = "python main.py --src imgs/{source_img} --dst imgs/{target_img} --out results/{out} --correct_color --warp_2d".format(
    source_img = source_name, target_img = target_name, out = result_name)
os.system(command)
```

## 4. Result Images

```
%cd /content/FaceSwap/results

# faceswap 된 이미지
result_image = Image(filename=result_name)
display(result_image)

%cd /content/FaceSwap/imgs

# source image
source_image = Image(filename=source_name)
display(source_image)

%cd /content/FaceSwap
```
